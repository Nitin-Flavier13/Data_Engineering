```python
import json
import boto3
from datetime import datetime
from io import StringIO
import pandas as pd

def album(data):
    album_list = []
    for row in data['items']:
        album_id = row['track']['album']['id']
        album_name = row['track']['album']['name']
        album_release_date = row['track']['album']['release_date']
        album_total_tracks = row['track']['album']['total_tracks']
        album_external_url = row['track']['album']['external_urls']['spotify']
        album_dic = {"album_id":album_id,"name":album_name,"release_date":album_release_date,"total_tracks":album_total_tracks,"url":album_external_url}
        album_list.append(album_dic)
    
    return album_list

def  artist(data):
    # going for the artists
    artist_list = []
    c=1
    print(data['items'][2]['track']['artists'])
    for row in data['items']:
        for artist in row['track']['artists']:
            artist_id = artist['id']
            artist_name = artist['name']
            artist_url = artist['external_urls']['spotify']
            artist_dic = {'id':artist_id,'artist_no':c,'name':artist_name,'url':artist_url}
            artist_list.append(artist_dic)
            c = c+1
    return artist_list

def song(data):
    # song list
    song_list = []
    # print(data['items'][0]['track'])
    for row in data['items']:
        song_id = row['track']['id']
        song_name = row['track']['name']
        song_duration = row['track']['duration_ms']
        song_url = row['track']['external_urls']['spotify']
        song_popularity = row['track']['popularity']
        song_added_at = row['added_at']
        album_id = row['track']['album']['id']
        for artist in row['track']['artists']:
            artist_id = artist['id']
            song_dic = {'id':song_id,'name':song_name,"duration":song_duration,"url":song_url,"popularity":song_popularity,"added_at":song_added_at,"album_id":album_id,"artist_id":artist_id}
            song_list.append(song_dic)
    
    return song_list
    

    
s3 = boto3.client('s3')

# val = s3.list_objects(
#     Bucket = "etl-pipeline-nitinflv",
#     Prefix = "raw_data/to_be_processed/"
# )
# print(val)
bucket_name = "etl-pipeline-nitinflv"

content = s3.list_objects(
    Bucket = bucket_name,
    Prefix = "raw_data/to_be_processed/"
)['Contents']
# print(content)

spotify_data = []
spotify_keys = []
    
for file in content:
    file_key = file['Key']
    # print(file_key)
    if file_key.split('.')[-1] == 'json':
        meta_data = s3.get_object(Bucket = bucket_name,Key = file_key)
        content = meta_data['Body']
        pythonObject = json.loads(content.read())     # json.loads deserializes JSON formatted string
        # print(pythonObject)                           #  an then convert to python object 
        spotify_data.append(pythonObject)
        spotify_keys.append(file_key)
        print(type(pythonObject))
            
    
for data in spotify_data:
    album_list = album(data)
    song_list  = song(data)
    artist_list = artist(data)
    
    print(album_list)
    
    album_df = pd.DataFrame.from_dict(album_list)
    artist_df = pd.DataFrame.from_dict(artist_list)
    song_df = pd.DataFrame.from_dict(song_list)
    
    print(album_df)
    
    album_df = album_df.drop_duplicates(subset=['album_id'])
    artist_df = artist_df.drop_duplicates(subset=['id','name'])
    song_df = song_df.drop_duplicates(subset=['id','name','artist_id'])
    
    
    # album_df['release_date'] = pd.to_datetime(album_df['release_date'])
    album_df['release_date'] = pd.to_datetime(album_df['release_date'], errors='coerce', format='%Y-%m-%d')
    song_df['added_at'] = pd.to_datetime(song_df['added_at'])
    
    song_key = "transformed_data/song_data/song_transformed_" + str(datetime.now()) + ".csv"
    album_key = "transformed_data/album_data/album_transformed_" + str(datetime.now()) + ".csv"
    artist_key = "transformed_data/artist_data/artist_transformed_" + str(datetime.now()) + ".csv"
        
        
    song_buffer = StringIO()
    song_df.to_csv(song_buffer,index = False)
    song_content = song_buffer.getvalue()
    
    s3.put_object(
        Bucket = bucket_name,
        Key = song_key,
        Body = song_content
    )
    
    album_buffer = StringIO()
    album_df.to_csv(album_buffer,index = False)
    album_content = album_buffer.getvalue()
    
    s3.put_object(
        Bucket = bucket_name,
        Key = album_key,
        Body = album_content
    )
    
    artist_buffer = StringIO()
    artist_df.to_csv(artist_buffer,index = False)
    artist_content = artist_buffer.getvalue()
        
    s3.put_object(
        Bucket = bucket_name,
        Key = artist_key,
        Body = artist_content
    )
        
s3_resource = boto3.resource('s3')
for key in spotify_keys:
    copy_source = {
        'Bucket' : bucket_name,
        'Key' : key
    }
    s3_resource.meta.client.copy(copy_source,bucket_name,'raw_data/processed/' + key.split('/')[-1])
    s3_resource.Object(bucket_name,key).delete()
        
        
```

```python
import boto3

def create_glue_crawler(crawler_name, s3_path, database_name, role_name):
    glue_client = boto3.client('glue')

    # Use existing IAM role
    role_arn = "arn:aws:iam::891377255085:role/service-role/AWSGlueServiceRole-spotify"

    # Create or update Glue Crawler
    response = glue_client.create_crawler(
        Name=crawler_name,
        Role=role_arn,  # Use the existing IAM role ARN
        DatabaseName=database_name,
        Targets={
            'S3Targets': [
                {
                    'Path': s3_path
                },
            ]
        },
        SchemaChangePolicy={
            'UpdateBehavior': 'UPDATE_IN_DATABASE',
            'DeleteBehavior': 'DELETE_FROM_DATABASE'
        },
        RecrawlPolicy={
            'RecrawlBehavior': 'CRAWL_EVERYTHING'
        }
    )

    print(f"Crawler {crawler_name} created successfully!")


crawler_name = 'SpotifyCrawler'
s3_path = 's3://etl-pipeline-nitinflv/transformed_data/'
database_name = 'db_spotify'
role_name = 'AWSGlueServiceRole-spotify'

create_glue_crawler(crawler_name, s3_path, database_name, role_name)
```

```python
# run crawler

import boto3
import time

def run_glue_crawler(crawler_name):
    glue_client = boto3.client('glue')

    # Start the Glue Crawler
    response = glue_client.start_crawler(
        Name=crawler_name
    )

    print(f"Starting Glue Crawler {crawler_name}...")

    # Poll for the status of the crawler
    while True:
        response = glue_client.get_crawler(Name=crawler_name)
        crawler_status = response['Crawler']['State']

        print(f"Crawler {crawler_name} status: {crawler_status}")

        if crawler_status == 'READY':
            print(f"Glue Crawler {crawler_name} completed successfully!")
            break
        elif crawler_status in ['STOPPING', 'FAILED', 'STOPPED']:
            print(f"Glue Crawler {crawler_name} failed or stopped.")
            break

        time.sleep(30)  


crawler_name = 'SpotifyCrawler' 
run_glue_crawler(crawler_name)
```
